```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification", device=0) # to utilize GPU
```

```python
candidate_labels = ["History", "Biology", "Maths", "Coding"]
sent = input("Enter question: ")
res = classifier(sent, candidate_labels)
subject_scores = {res['labels'][i]:res['scores'][i] for i in range(4) if res['scores'][i]>=0.5 }
if len(subject_scores)==0:
  print('Couldnt Classify')
else:
  print(list(subject_scores.keys())[0])
```
